```python
#import tensorflow as tf
import cv2
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #graph
from tensorflow .keras.models import Sequential #ANN architecture
from tensorflow .keras.layers import Dense #the layers in the ANN
from tensorflow .keras.utils import to_categorical
```

```python
data=!unzip /content/digits-20200922T072401Z-001.zip
```

```python
data
```

```python
t=[]
r=[]
#r1=[]
data1="digits/digit_"
for i in range(10):
    j=str(i)
    files=glob.glob(data1+j+"/*")
    
    n=len(files)
    print(n)
    for k in range(n):
        img=cv2.imread(files[k])
        #print(img.shape)
        t.append(img.flatten())
        r.append(i)
        #r1.append(i)
```

```python
#normalize the images.normalize the pixel values from [0,255] tp
#[-0.5 - 0.5] to make our network easier to train
t=[x/255 for x in t]
```

```python
#result vector
r=to_categorical(r) #convert to categorical
r
```

```python
X=np.array(t) 
y=np.array(r)
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, train_size=0.75)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#build the model
model=Sequential()
model.add(Dense(32,activation="relu", input_dim=3072))
model.add(Dense(10,activation="softmax"))
```

```python
#compile model
#the loss function measures how well the model did on training and tries to
#improve on it using optimizer
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy" , #(classes that are greater than 2)
    metrics=["accuracy"]
)
```

```python
#train the model
model.fit(X_train,y_train,epochs=100,batch_size=32)
```

```python
model.evaluate(X_test,y_test)
```

```python
import time
```

```python
nodes=[8,16,32,64,128,256,512,1024]
train_acc=[]
test_acc=[]
times=[]
```

```python

def diff_nodes(i):
  model=Sequential()
  model.add(Dense(i,activation="relu", input_dim=3072))
  model.add(Dense(10,activation="softmax"))
  model.compile(optimizer="adam",loss="categorical_crossentropy" , metrics=["accuracy"])

  t_s=time.time()
  model.fit(X_train,y_train,epochs=100,batch_size=32)
  t_e=time.time()
  times.append(t_e-t_s)

  a=model.evaluate(X_train,y_train)
  b=model.evaluate(X_test,y_test)

  train_acc.append(a[1])
  test_acc.append(b[1])

```

```python
for j in nodes:
  diff_nodes(j)
```

```python
print(times)
print(train_acc)
print(test_acc)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=nodes,y=train_acc)
plt.xlabel('Nodes')
plt.ylabel('Train Accuracy')
plt.plot()
```

```python
sns.barplot(x=nodes,y=test_acc)
plt.xlabel('Nodes')
plt.ylabel('Test Accuracy')
plt.plot()
```

```python
sns.barplot(x=nodes,y=times)
plt.xlabel('Nodes')
plt.ylabel('Time')
plt.plot()
```
